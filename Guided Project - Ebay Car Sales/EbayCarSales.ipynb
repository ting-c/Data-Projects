# Ebay Car Sales

In this project, we'll work with a modified dataset of used cars from eBay Kleinanzeigen, a classifieds section of the German eBay website. The original dataset was obtained from [Kaggle](https://www.kaggle.com/orgesleka/used-cars-database/data.). The modified dataset took a sampled of  50,000 data points from the full dataset and modified a bit to more closely resemble what we would expect from a scraped dataset. The aim of this project is to clean the data and analyze the included used car listings. 

The data dictionary provided with data is as follows:

- dateCrawled - When this ad was first crawled. All field-values are taken from this date.
- name - Name of the car.
- seller - Whether the seller is private or a dealer.
- offerType - The type of listing
- price - The price on the ad to sell the car.
- abtest - Whether the listing is included in an A/B test.
- vehicleType - The vehicle Type.
- yearOfRegistration - The year in which the car was first registered.
- gearbox - The transmission type.
- powerPS - The power of the car in PS.
- model - The car model name.
- kilometer - How many kilometers the car has driven.
- monthOfRegistration - The month in which the car was first registered.
- fuelType - What type of fuel the car uses.
- brand - The brand of the car.
- notRepairedDamage - If the car has a damage which is not yet repaired.
- dateCreated - The date on which the eBay listing was created.
- nrOfPictures - The number of pictures in the ad.
- postalCode - The postal code for the location of the vehicle.
- lastSeenOnline - When the crawler saw this ad last online.

### Variations across different car brands

We're going to look at the the top 20 brands of car in the dataset.

```python
#frequency table with proportion
top_20_brand = autos['brand'].value_counts(normalize = True ,dropna = False).sort_values(ascending = False).head(20)
top_20_brand
```

The table above shows that volkswagen dominated the ebay site with more than 20% of the dateset while bmw , opel , mercedes_benz  and audi represent above 40 % together. We'll focus on these top 5 brands for the next section of our analysis.

```python
top_5_brand = ['volkswagen' , 'bmw', 'opel', 'mercedes_benz', 'audi']
top_5_brand
```

### Average Price

```python
mean_price_all_brand= {}
for brand in top_5_brand:
    selected_rows = autos[autos['brand'] == brand]    
    brand_price = selected_rows['price']
    
    total_brand_price = 0
    for price in brand_price:
        total_brand_price += price
    mean_brand = round((total_brand_price / len(brand_price)),2)
    mean_price_all_brand[brand] = mean_brand    
    
print(mean_price_all_brand)
```

```python
# Displaying the dict as in a table
price_series = pd.Series(mean_price_all_brand)
mean_price = pd.DataFrame(price_series, columns=["mean_price"])
mean_price
```

Audi has the highest mean advertised selling price while Opel has the lowest selling price. 

### Average Odometer

```python
mean_odometer_all_brand= {}
for brand in top_5_brand:
    selected_rows = autos[autos['brand'] == brand]    
    brand_odometer = selected_rows['odometer_km']
    
    total_brand_odometer = 0
    for odometer in brand_odometer:
        total_brand_odometer += odometer
    mean_odometer = round(total_brand_odometer / len(brand_odometer))
    mean_odometer_all_brand[brand] = mean_odometer    
    
print(mean_odometer_all_brand)
```

```python
odo_series = pd.Series(mean_odometer_all_brand)
odometer_info = pd.DataFrame(odo_series, columns=["mean_odometer_in_km"])
odometer_info
```

The mean odometer is very uniform across all top 5 brands which could mean that Ebay sites tends to have cars over 129,000 km mileage on sale.

### Average Registration Year

```python
mean_reg_all_brand= {}
for brand in top_5_brand:
    selected_rows = autos[autos['brand'] == brand]    
    brand_reg = selected_rows['registration_year']
    
    total_brand_reg = 0
    for reg in brand_reg:
        total_brand_reg += reg
    mean_reg = round((total_brand_reg / len(brand_reg)),2)
    mean_reg_all_brand[brand] = mean_reg    
    
print(mean_reg_all_brand)
```

```python
reg_series = pd.Series(mean_reg_all_brand)
reg_info = pd.DataFrame(reg_series, columns=["mean_registration_year"])
reg_info
```

The average registration year for the top 5 brands are similar with Audi cars tends to be newer and mercedes_benz cars tends to be older.
