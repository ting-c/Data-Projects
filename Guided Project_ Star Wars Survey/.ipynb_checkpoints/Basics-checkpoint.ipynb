```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding="ISO-8859-1")
star_wars.columns
```

```python
star_wars.head()
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars['RespondentID']
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
## Converting to boolean types
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
# Convert NaN to False , response to True
mapping_dict = {
    "Star Wars: Episode I  The Phantom Menace": True,    
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False,
   }

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(mapping_dict)

star_wars[3:20]
```

```python
# Rename columns
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
# Change the columns to float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars.columns[9:15].unique()
```

```python
# Rename with more descriptive name
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.columns[9:15]
```

```python
# Compute the mean of each ranking columns

star_wars[star_wars.columns[9:15]].mean()
```

```python
# Plot bar chart
%matplotlib inline
star_wars[star_wars.columns[9:15]].mean().plot(kind='bar')
```

The columns names have been cleaned and the bar chart above shows the ranking for each movie. The last three are ranked higher on average than the previous three.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot(kind='bar')
```

It seems like the movies that are ranked higher on average are also being seen more by people.

```python
# examine how certain segments of the survey population responded

star_wars_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
star_wars_not_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]
```

```python
# Redo Analysis

print(star_wars_fan[star_wars_fan.columns[9:15]].mean())
print(star_wars_not_fan[star_wars_not_fan.columns[9:15]].mean())
```

```python
star_wars_fan[star_wars_fan.columns[9:15]].mean().plot(kind='bar')
```

```python
star_wars_not_fan[star_wars_not_fan.columns[9:15]].mean().plot(kind='bar')
```

The average ranking for all non fans are actually higher than fan ranking. The ranking for episode III was also the lowest for both group which could indicate that the movie was poorly made and didn't appeal to audiences.
