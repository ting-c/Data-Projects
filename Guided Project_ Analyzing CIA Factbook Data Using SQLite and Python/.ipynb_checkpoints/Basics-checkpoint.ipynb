```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('factbook.db')
cursor = conn.cursor()
# Query the database to get the name of the table and what the table looked like
query = "SELECT * FROM sqlite_master WHERE type='table';"
# Execute before fetchall rows
cursor.execute(query)
cursor.fetchall()
```

```python
# To return the results in a table
pd.read_sql_query(query, conn)
```

```python
# Run query to return first 5 rows of facts table
query_fact = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(query_fact, conn)
```

### Summary Statistics

```python
q1 = "select MIN(population) 'Minimum population',MAX(population) 'Maximum population', MIN(population_growth) 'Minimum population growth', MAX(population_growth) 'Maximum population growth' FROM facts"
pd.read_sql_query(q1, conn)
```

### Population of 0

```python
q2 = "SELECT * FROM facts WHERE population = 0"
pd.read_sql_query(q2, conn)
```

### Max population

```python
q3 = "SELECT * FROM facts WHERE population = (SELECT MAX(population) from facts)"
pd.read_sql_query(q3, conn)
```

It appears that there is a row for the whole world and another row for Antartica which contains population of 0. [CIA Factbook](https://www.cia.gov/library/publications/the-world-factbook/geos/ay.html)

### Histogram 

#### removing Antartica and World rows

```python
q4 = "SELECT population,population_growth,birth_rate,death_rate FROM facts WHERE 0 < population and population < (SELECT MAX(population) from facts)"
pd.read_sql_query(q4, conn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1)
pd.read_sql_query(q4, conn).hist(ax=ax)
```

### Countries with highest population density

```python
q5 = "SELECT name Country, CAST(population as float)/CAST(area_land as float) 'Population Density' from facts ORDER BY CAST(population as float)/CAST(area_land as float) DESC"
pd.read_sql_query(q5, conn)
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1)
pd.read_sql_query(q5, conn).hist(ax=ax)
```

### Top countries with highest ratios of water to land

```python
q6 = "SELECT name Country, CAST(area_water as float)/CAST(area_land as float) 'Water to Land Ratio' from facts ORDER BY CAST(area_water as float)/CAST(area_land as float) DESC"
pd.read_sql_query(q6, conn)
```

British Indian Ocean Territory have significantly more water area to land area compared to rest of the world.
